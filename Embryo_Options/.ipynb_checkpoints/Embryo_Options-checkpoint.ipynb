### Question 5

*Q: Identify all records where the 'Billing Month' does not match the month of the 'Date In'.*

A: Create two new columns that contain the numeric value of the months from both original columns (*'Billing Month'* and *'Date In'*). Do this by converting each column to datetime in some form and pulling the **month** information. Finally, find the rows where the values in the two newly created columns *don't* match.

```python
from datetime import datetime
```

```python
# Check for null values
charges[['Date In', 'Billing Month']].isnull().sum()
```

```python
# Replace all null cells with a string ('?') and leave all other cells as they are
charges['Billing Month'] = charges['Billing Month'].map(lambda x: '?' if str(x) == 'nan' else str(x))
charges['Date In'] = charges['Date In'].map(lambda x: '?' if str(x) == 'nan' else x)
```

```python
# Create function that will yield the corresponding number to the 'Billing Month'
def month1(x):
    d = datetime.strptime(x, '%B')
    if x != '?':
        return int(d.strftime('%m'))
    else:
        return x
```

```python
# Create function that will yield the corresponding number to the 'Date In'
def month2(x):
    try: # Using try/except here in case there are any weird strings that don't fit the format I define below
        d = datetime.strptime(x, '%m/%d/%y')
        if x != '?':
            return int(d.strftime('%m'))
    except:
        return x
```

```python
# Create new column with 'Billing Month' converted into corresponding number (int) if doesn't equal '?'
# Using month1 function from above
charges['Bill_Month_num'] = charges['Billing Month'].map(lambda x: month1(x) if x != '?' else x)
```

```python
# Create new column with 'Date In' converted into corresponding number (int) if doesn't equal '?'
# Using month2 function from above
charges['Date_In_num'] = charges['Date In'].map(lambda x: month2(x) if x != '?' else x)
```

```python
# First 30 rows - noticing that there are cells in the Date_In_num column that did not get converted because there is >1 date in the string
charges[['Bill_Month_num', 'Date_In_num']].head(30)
```

```python
# Find the rows where the month values don't match
non_match = charges[['Bill_Month_num', 'Date_In_num']][charges['Bill_Month_num'] != charges['Date_In_num']]
```

```python
# Find how many rows out of the whole dataset have mismatching months
len(non_match)
```

```python
# Show records where 'Billing Month' does not match the month of the 'Date In'
non_match.head(10)
# non_match
```

After cleaning the data to an extent, I've found two patterns concerning the rows containing months that "don't match":

1. `'Bill_Month_num'` = `'?'` when `'Date_In_num'` is some integer (**ex**: `charges.iloc[26]`)
2. `'Bill_Month_num'` is some integer when `'Date_In_num'` remains a string that contains *two* dates (**ex**: `charges.iloc[43]`)

That being said, you could clean/manipulate the `'Date_In_num'` values even further to separate the multiple date situations and write a function to see if the month in *either* date matched the `'Bill_Month_num'` value in each case. 

It would obviously take additional research to figure out why some of the `'Bill_Month_num'` rows contained nulls (i.e. '?').
