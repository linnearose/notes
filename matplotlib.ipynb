### barplots in seaborn | 3.19.19

```python
import numpy as np
from numpy import median, mean, random
from random import randint

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load preexisting seaborn dataset
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x='day', y='tip', data=tips);
```

```python
sns.barplot(x='day', y='total_bill', data=tips);
```

```python
sns.barplot(x='day', 
            y='tip', 
            data=tips, 
            hue='sex', 
            palette='winter_r');
```

```python
# Hue, palette
sns.barplot(x='day', 
            y='total_bill', 
            data=tips, 
            hue='smoker');
```

```python
# Horizontal bar plot
sns.barplot(x='total_bill', 
            y='day', 
            data=tips, 
            palette='vlag');
```

```python
# Order labels
sns.barplot(x='day', 
            y='tip', 
            data=tips, 
            palette='spring', 
            order=['Sat', 'Fri', 'Sun', 'Thur']);
```

```python
# Change estimator (mean, median, etc.)
sns.barplot(x='day', 
            y='total_bill', 
            data=tips,
            estimator=median, # Default is mean
            palette='spring');
```

```python
sns.barplot(x='smoker', 
            y='tip', 
            data=tips,
            estimator=median,
            hue='sex',
            palette='coolwarm');
```

```python
sns.barplot(x='smoker', 
            y='tip', 
            data=tips,
            estimator=mean,
            hue='sex',
            palette='winter_r',
            saturation=0.5);
```

```python
# Try out randomly generated palettes
def rand_palette(palette_list):
    r = randint(0, len(palette_list))
    print('palette name: ', palette_list[r])
    
    sns.barplot(x='day', 
                y='total_bill', 
                data=tips, 
                palette=palette_list[r]);
    
```
